```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/pivot_table_of_results.csv')
```

```python
# For the readout specifically for voltage in read-out coloumn, change the measurement_name column names Pulse Width 10%, 50% and 90% to APD10, APD50 and APD90
df.loc[df['read-out'] == 'voltage', 'measurement_name'] = df.loc[df['read-out'] == 'voltage', 'measurement_name'].replace({
    'Pulse Width 10%': 'APD10',
    'Pulse Width 50%': 'APD50',
    'Pulse Width 90%': 'APD90'
})
```
